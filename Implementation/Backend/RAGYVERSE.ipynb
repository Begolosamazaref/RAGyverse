```python
import os
import tempfile
import soundfile as sf
import ffmpeg
import json
import fitz
import getpass
from vosk import Model, KaldiRecognizer
from langchain.document_loaders import PyPDFLoader
from langchain.docstore.document import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.prompts import PromptTemplate
from langchain.chat_models import init_chat_model
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
import numpy as np
from pydub import AudioSegment
from gtts import gTTS
from transformers import pipeline
from flask import Flask, request, jsonify, send_file
import threading
import time
from pyngrok import conf, ngrok
from flask_cors import CORS

# --- Global Setup ---
app = Flask(__name__)
CORS(app, resources={r"/*": {"origins": "*"}})  # Allow all origins for all routes

# Securely set API key for Groq
if not os.environ.get("GROQ_API_KEY"):
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter API key for Groq: ")

# Load Vosk ASR Model
vosk_model_path = "vosk_model/vosk-model-en-us-0.22"
model_vosk = Model(vosk_model_path)

# Load SmolDocling OCR Model
ocr_pipe = pipeline("image-text-to-text", model="ds4sd/SmolDocling-256M-preview")

# Global LangChain components
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)
retriever = None
conv_chain = None
document_text = ""

# --- Helper Functions ---
def transcribe_audio(audio_path):
    """Transcribes speech to text using Vosk."""
    try:
        with sf.SoundFile(audio_path) as audio_file:
            recognizer = KaldiRecognizer(model_vosk, audio_file.samplerate)
            recognizer.SetWords(True)
            while True:
                data = audio_file.read(dtype="int16")
                if len(data) == 0:
                    break
                recognizer.AcceptWaveform(data.tobytes())
            result = json.loads(recognizer.FinalResult())
            return result.get("text", "⚠️ Could not transcribe audio")
    except Exception as e:
        print("Error in transcribe_audio:", e)
        return f"❌ Error during transcription: {e}"

def convert_audio_to_wav(file_storage):
    """Converts audio to WAV format from a FileStorage object."""
    try:
        # Save the FileStorage object to a temporary file
        with tempfile.NamedTemporaryFile(delete=False, suffix=".tmp") as temp_file:
            file_storage.save(temp_file.name)
            temp_path = temp_file.name

        # Convert the temporary file to WAV
        temp_wav = tempfile.NamedTemporaryFile(delete=False, suffix=".wav").name
        ffmpeg.input(temp_path).output(temp_wav, format="wav", ar=16000, ac=1).run(overwrite_output=True)

        # Clean up the original temporary file
        os.unlink(temp_path)

        return temp_wav
    except Exception as e:
        print("Error in convert_audio_to_wav:", e)
        return f"❌ Audio conversion failed: {e}"

def extract_text(file_storage):
    """Extracts text from an uploaded PDF file using OCR when necessary."""
    try:
        # Save the uploaded file to a temporary location
        with tempfile.NamedTemporaryFile(delete=False, suffix=".pdf") as temp_file:
            file_storage.save(temp_file.name)
            pdf_path = temp_file.name

        # Open the PDF using fitz
        doc = fitz.open(pdf_path)
        extracted_text = []

        for page in doc:
            text = page.get_text("text")
            if text.strip():
                extracted_text.append(text)
            else:
                images = page.get_images(full=True)
                for img in images:
                    xref = img[0]
                    base_image = doc.extract_image(xref)
                    image_bytes = base_image["image"]
                    messages = [{"role": "user", "content": image_bytes}]
                    ocr_result = ocr_pipe(messages)
                    if ocr_result:
                        extracted_text.append(ocr_result[0]['generated_text'])

        # Clean up: remove the temporary file
        os.unlink(pdf_path)

        return "\n".join(extracted_text) if extracted_text else "⚠️ No extractable text found."

    except Exception as e:
        print("Error in extract_text:", e)
        return f"❌ Error processing file: {e}"

def text_to_speech(text, filename):
    """Converts text to speech and saves it as an MP3 file."""
    try:
        audio_path = os.path.join("static", f"{filename}.mp3")
        os.makedirs("static", exist_ok=True)
        print(f"Generating TTS audio at: {audio_path}")  # Debug: log the path
        tts = gTTS(text=text, lang='en')
        tts.save(audio_path)
        if os.path.exists(audio_path):
            print(f"Audio file successfully created at: {audio_path}")
        else:
            print(f"Audio file NOT found at: {audio_path}")
        return audio_path
    except Exception as e:
        print("Error in text_to_speech:", e)
        return f"❌ Error generating speech: {e}"



def initialize_conversational_chain(document_text):
    """Initializes a Conversational Retrieval Chain using the document text."""
    global retriever, conv_chain
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    docs = text_splitter.split_documents([Document(page_content=document_text)])
    vectorstore = FAISS.from_documents(docs, embeddings)
    retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
    llm_model = init_chat_model("llama3-8b-8192", model_provider="groq")
    prompt_template = PromptTemplate(
        template="""You are a highly knowledgeable multilingual AI tutor skilled in conversational pedagogy.
        Context: {context}
        Question: {question}
        Answer: """,
        input_variables=["context", "question"]
    )
    conv_chain = ConversationalRetrievalChain.from_llm(
        llm=llm_model,
        retriever=retriever,
        memory=memory,
        combine_docs_chain_kwargs={"prompt": prompt_template}
    )
    return "✅ Conversational Chain Initialized"

def chat_with_bot(user_message):
    """Conversational interface: Uses the conversational chain to respond to user messages."""
    global conv_chain, document_text
    if not document_text:
        return {"error": "No PDF processed. Please upload a PDF first."}, None
    if not conv_chain:
        result = initialize_conversational_chain(document_text)
        if "❌" in result or "⚠️" in result:
            return {"error": result}, None
    try:
        response = conv_chain({"question": user_message})
        answer = response.get("answer", "No answer returned.")
        audio_path = text_to_speech(answer, f"response_{int(time.time())}")
        print(f"Chat response: answer={answer}, audio_path={audio_path}")  # Debug
        return {"answer": answer}, audio_path
    except Exception as e:
        print("Error in chat_with_bot:", e)
        return {"error": f"❌ Error in chat_with_bot: {e}"}, None

# --- Flask API Routes ---
@app.route('/upload_pdf', methods=['POST'])
def upload_pdf():
    global document_text
    if 'file' not in request.files:
        return jsonify({"error": "No file part"}), 400
    file = request.files['file']
    if file.filename == '':
        return jsonify({"error": "No selected file"}), 400
    if not file.filename.endswith('.pdf'):
        return jsonify({"error": "Invalid file format. Please upload a PDF."}), 400

    document_text = extract_text(file)
    if document_text.startswith("❌") or document_text.startswith("⚠️"):
        return jsonify({"error": document_text}), 500
    return jsonify({"message": "PDF processed successfully", "text": document_text}), 200

@app.route('/ask_text', methods=['POST'])
def ask_text():
    data = request.json
    if not data or 'question' not in data:
        return jsonify({"error": "No question provided"}), 400
    question = data['question']
    response, audio_path = chat_with_bot(question)
    if 'error' in response:
        return jsonify(response), 500
    audio_filename = os.path.basename(audio_path).replace('.mp3', '')
    audio_url = f"/tts/{audio_filename}"
    print(f"Returning audio URL: {audio_url}")  # Debug
    return jsonify({"answer": response['answer'], "audio_url": audio_url}), 200

@app.route('/ask_audio1', methods=['POST'])
def ask_audio1():
    if 'file' not in request.files:
        return jsonify({"error": "No audio file part"}), 400
    audio_file = request.files['file']
    if audio_file.filename == '':
        return jsonify({"error": "No selected audio file"}), 400
    temp_audio_path = convert_audio_to_wav(audio_file)
    if temp_audio_path.startswith("❌"):
        return jsonify({"error": temp_audio_path}), 500
    question = transcribe_audio(temp_audio_path)
    if question.startswith("❌") or question.startswith("⚠️"):
        return jsonify({"error": question}), 500
    response, _ = chat_with_bot(question)  # Ignore the audio_path return value
    if 'error' in response:
        return jsonify(response), 500
    return jsonify({"question": question, "answer": response['answer']}), 200


@app.route('/ask_audio', methods=['POST'])
def ask_audio():
    if 'file' not in request.files:
        return jsonify({"error": "No audio file part"}), 400
    audio_file = request.files['file']
    if audio_file.filename == '':
        return jsonify({"error": "No selected audio file"}), 400
    temp_audio_path = convert_audio_to_wav(audio_file)
    if temp_audio_path.startswith("❌"):
        return jsonify({"error": temp_audio_path}), 500
    question = transcribe_audio(temp_audio_path)
    if question.startswith("❌") or question.startswith("⚠️"):
        return jsonify({"error": question}), 500
    response, audio_path = chat_with_bot(question)
    if 'error' in response:
        return jsonify(response), 500
    audio_filename = os.path.basename(audio_path).replace('.mp3', '')
    audio_url = f"/tts/{audio_filename}"
    print(f"Returning audio URL: {audio_url}")  # Debug
    return jsonify({"answer": response['answer'], "audio_url": audio_url}), 200


@app.route('/tts/<filename>', methods=['GET'])
def get_tts(filename):
    audio_path = os.path.join("static", f"{filename}.mp3")
    print(f"Requesting audio file: {audio_path}")  # Debug
    if not os.path.exists(audio_path):
        print(f"Audio file not found: {audio_path}")
        return jsonify({"error": "Audio file not found"}), 404
    return send_file(audio_path, mimetype="audio/mpeg")

@app.route('/reset_session', methods=['POST'])
def reset_session():
    global memory, conv_chain, retriever
    memory.clear()
    conv_chain = None
    retriever = None
    return jsonify({"message": "Session reset successfully"}), 200

if __name__ == "__main__":
    # Set your ngrok authtoken
    NGROK_AUTH_TOKEN = "2txkT81lzhDYaHOORvxgxS3NQ9x_66XDSE86P7wsnEn36sKrD"
    conf.get_default().auth_token = NGROK_AUTH_TOKEN

    def run_flask():
        app.run(host='0.0.0.0', port=5000)

    public_url = ngrok.connect(5000)
    print(f"Public URL: {public_url}")

    flask_thread = threading.Thread(target=run_flask, daemon=True)
    flask_thread.start()

    while True:
        time.sleep(1)
```
